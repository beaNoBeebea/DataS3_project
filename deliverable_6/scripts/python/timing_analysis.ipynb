```python
import matplotlib.pyplot as plt
import numpy as np 

xpoints = np.array([10000, 50000, 100000, 500000, 1000000])
ypoints_without = np.array([0.110, 0.125, 0.500, 2.735, 18.219])
ypoints_with = np.array([0.031, 0.125, 0.172, 0.687, 3.078])

plt.plot(xpoints, ypoints_without, label='Without Index', linewidth=2, markersize=8)
plt.plot(xpoints, ypoints_with, label='With Index', linewidth=2, markersize=8)

plt.xlabel('Number of Rows', fontsize=12)
plt.ylabel('Execution Time (seconds)', fontsize=12)
plt.title('Impact of Indexing on Query Performance', fontsize=14)

plt.xticks(xpoints, ['10k', '50k', '100k', '500k', '1M'])

plt.tight_layout()

plt.show()
```
